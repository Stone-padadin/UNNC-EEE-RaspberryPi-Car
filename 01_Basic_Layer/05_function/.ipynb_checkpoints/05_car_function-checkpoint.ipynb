```python
# ====================== 第五课：Python函数封装 + 小车智能行驶 ======================
# 核心目标：将小车校验/执行/统计逻辑封装为函数，实现代码复用与整洁

# ---------------------- 1. 全局配置常量（大写区分普通变量） ----------------------
MAX_SPEED = 50    # 最大限速
MIN_SPEED = 8     # 最低速度
VALID_DIRS = ["forward", "backward", "left", "right"]  # 合法方向

# ---------------------- 2. 核心功能函数定义 ----------------------
# 函数1：速度校验（输入原始速度，返回修正后速度）
def check_speed(original_speed):
    """
    速度校验函数：修正超标/过低的速度
    :param original_speed: 原始速度（数值型）
    :return: 修正后的速度（数值型）
    """
    if original_speed > MAX_SPEED:
        print(f"⚠️ 速度超标（{original_speed}），修正为{MAX_SPEED}")
        return MAX_SPEED
    elif original_speed < MIN_SPEED:
        print(f"⚠️ 速度过低（{original_speed}），修正为{MIN_SPEED}")
        return MIN_SPEED
    return original_speed

# 函数2：方向校验（输入方向，返回是否合法）
def check_direction(direction):
    """
    方向校验函数：判断方向是否在合法列表中
    :param direction: 行驶方向（字符串）
    :return: 合法返回True，非法返回False
    """
    if direction not in VALID_DIRS:
        print(f"❌ 方向非法（{direction}）")
        return False
    return True

# 函数3：开灯判断（输入速度+方向，返回是否开灯）
def judge_light(speed, direction):
    """
    开灯规则：速度≥30且正向行驶则开灯
    :param speed: 修正后的速度
    :param direction: 合法方向
    :return: 开灯返回True，否则返回False
    """
    if speed >= 30 and direction == "forward":
        print("✅ 速度≥30且正向行驶，打开车灯")
        return True
    return False

# 函数4：单段路径执行（整合所有校验，返回有效数据）
def run_single_path(path):
    """
    执行单段路径：整合距离/方向/速度校验，返回有效数据或None
    :param path: 单段路径字典（speed/direction/distance）
    :return: 有效数据字典（distance/speed/direction/light），非法返回None
    """
    # 提取参数
    speed = path["speed"]
    direction = path["direction"]
    distance = path["distance"]
    
    # 距离校验
    if distance <= 0:
        print("❌ 距离非法（≤0），跳过该段")
        return None
    
    # 方向校验（调用子函数）
    if not check_direction(direction):
        return None
    
    # 速度校验（调用子函数）
    corrected_speed = check_speed(speed)
    
    # 开灯判断（调用子函数）
    light_on = judge_light(corrected_speed, direction)
    
    # 返回有效数据
    return {
        "distance": distance,
        "speed": corrected_speed,
        "direction": direction,
        "light": light_on
    }

# 函数5：数据统计（输入有效路径，返回统计结果）
def calculate_stats(valid_paths):
    """
    统计总距离/正向距离/开灯次数/最高/平均速度
    :param valid_paths: 有效路径列表
    :return: 统计结果字典
    """
    if not valid_paths:  # 空列表处理
        return {"total_distance":0, "forward_distance":0, "light_count":0, "max_speed":0, "avg_speed":0}
    
    total_dist = 0
    forward_dist = 0
    light_count = 0
    speed_list = []
    
    for path in valid_paths:
        total_dist += path["distance"]
        speed_list.append(path["speed"])
        if path["direction"] == "forward":
            forward_dist += path["distance"]
        if path["light"]:
            light_count += 1
    
    max_speed = max(speed_list)
    avg_speed = round(sum(speed_list)/len(speed_list), 1)
    
    return {
        "total_distance": total_dist,
        "forward_distance": forward_dist,
        "light_count": light_count,
        "max_speed": max_speed,
        "avg_speed": avg_speed
    }

# ---------------------- 3. 主程序：调用函数执行 ----------------------
if __name__ == "__main__":
    # 预设路径列表
    path_plan = [
        {"speed":28, "direction":"forward", "distance":12},
        {"speed":55, "direction":"right", "distance":6},
        {"speed":5, "direction":"forward", "distance":18},
        {"speed":40, "direction":"backward", "distance":-5},
        {"speed":35, "direction":"up", "distance":10},
        {"speed":32, "direction":"forward", "distance":15}
    ]
    
    # 执行所有路径
    print("🚗 小车开始智能行驶！")
    valid_paths = []  # 存储有效路径数据
    for idx, path in enumerate(path_plan):
        print(f"\n=== 执行第{idx+1}段路径 ===")
        single_result = run_single_path(path)
        if single_result:  # 路径有效则加入列表
            valid_paths.append(single_result)
            print(f"✅ 执行成功 | 方向：{single_result['direction']} | 速度：{single_result['speed']} | 距离：{single_result['distance']} | 车灯：{'开' if single_result['light'] else '关'}")
    
    # 统计结果
    stats = calculate_stats(valid_paths)
    print("\n================ 行驶完成 ================")
    print(f"📊 总有效距离：{stats['total_distance']} 米")
    print(f"📊 正向行驶距离：{stats['forward_distance']} 米")
    print(f"📊 开灯路段数：{stats['light_count']} 段")
    print(f"📊 最高速度：{stats['max_speed']} km/h")
    print(f"📊 平均速度：{stats['avg_speed']} km/h")
```
