```python
# 第四课：列表与字典 → 小车多数据管理
# ====================== 1. 列表实战：小车多档位调速 ======================
print("=== 列表实战：小车多档位调速 ===")
# 1.1 创建小车速度档位列表（起步→巡航→高速→减速）
speed_gears = [10, 25, 40, 30, 15]  # 5个档位的速度
print("原始速度档位：", speed_gears)

# 1.2 取单个档位（索引从0开始）
print("第3档巡航速度：", speed_gears[2])  # 取第3个元素（40）
print("最后1档减速速度：", speed_gears[-1])  # 倒数第1个元素（15）

# 1.3 新增档位（应急加速档）
speed_gears.append(50)
print("新增应急档后：", speed_gears)

# 1.4 循环遍历所有档位，模拟小车切换速度
print("\n小车依次切换档位：")
for index, speed in enumerate(speed_gears):
    # enumerate：同时获取索引和值，方便显示「第X档」
    print(f"第{index+1}档 → 速度：{speed} km/h")

# 1.5 切片：取常用档位（第2-4档）
common_gears = speed_gears[1:4]  # 从索引1到3（不包含4）
print("\n常用档位（2-4档）：", common_gears)

# ====================== 2. 字典实战：小车参数配置 ======================
print("\n=== 字典实战：小车参数配置 ===")
# 2.1 创建小车参数字典（不同类型数据：数字、字符串、布尔）
car_config = {
    "speed": 30,          # 当前速度
    "direction": "forward",# 行驶方向：forward/backward/left/right
    "mode": "auto",       # 模式：auto（自动）/manual（手动）
    "light": True,        # 车灯：True（开）/False（关）
    "max_speed": 50       # 最大限速
}
print("初始小车配置：", car_config)

# 2.2 取单个参数
print("当前行驶方向：", car_config["direction"])
print("车灯状态：", "开" if car_config["light"] else "关")

# 2.3 修改参数（小车切换为手动模式，减速）
car_config["mode"] = "manual"
car_config["speed"] = 20
print("\n修改后配置：", car_config)

# 2.4 遍历所有参数，打印配置清单
print("\n小车完整配置清单：")
for key, value in car_config.items():
    print(f"{key} → {value}")

# ====================== 3. 综合实战：列表+字典+循环 → 多段路径调速 ======================
print("\n=== 综合实战：小车多段路径自动调速 ===")
# 定义多段路径（每个路径是一个字典，存速度+方向）
path_plan = [
    {"speed": 20, "direction": "forward", "distance": 10},  # 第一段：前进10米，20km/h
    {"speed": 30, "direction": "right", "distance": 5},     # 第二段：右转5米，30km/h
    {"speed": 25, "direction": "forward", "distance": 15},  # 第三段：前进15米，25km/h
    {"speed": 15, "direction": "left", "distance": 5},      # 第四段：左转5米，15km/h
    {"speed": 10, "direction": "backward", "distance": 8}   # 第五段：后退8米，10km/h
]

# 循环执行每段路径
for i, path in enumerate(path_plan):
    print(f"\n执行第{i+1}段路径：")
    print(f"→ 方向：{path['direction']}")
    print(f"→ 速度：{path['speed']} km/h")
    print(f"→ 距离：{path['distance']} 米")
    
    # 限速检查（复用字典里的max_speed）
    if path["speed"] > car_config["max_speed"]:
        print(f"⚠️ 速度超标，强制降至{car_config['max_speed']} km/h")
```
