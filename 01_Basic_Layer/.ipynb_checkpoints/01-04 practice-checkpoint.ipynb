```python
#定义变量
max_speed = 50  #最大速度
min_speed = 8   #最低速度
valid_directions = ["forword", "backward", "left", "right"] #legel directions
```

```python
#统计变量
total_distance = 0  #总有效距离
speed_list = []     #有效速度列表
forward_distance = 0#正向有效距离
light_on_count = 0  #开灯路段数
```

```python
#定义预设路径
path_plan = [
    {"speed": 28, "direction": "forward", "distance": 12, "light": False},
    {"speed": 55, "direction": "right", "distance": 6, "light": False},
    {"speed": 5, "direction": "forward", "distance": 18, "light": False},
    {"speed": 40, "direction": "backward", "distance": -5, "light": False},  # 距离≤0
    {"speed": 35, "direction": "up", "distance": 10, "light": False},        # 非法方向
    {"speed": 32, "direction": "forward", "distance": 15, "light": False}
]
```

```python
print("🚗 小车开始智能行驶！")
# 遍历每段路径（你需要补全以下逻辑）
for idx, path in enumerate(path_plan):
    print(f"\n=== 执行第{idx+1}段路径 ===")
    # 1. 提取当前段参数（你写）
    current_speed = path["speed"]
    current_dir = path["direction"]
    current_dist = path["distance"]
    current_light = path["light"]
    
    # 2. 距离校验：≤0则跳过（你写）
    if  current_dist <=0:
        print("❌ 距离非法，跳过该段")
        continue
    
    # 3. 方向校验：非法则跳过（你写）
    if  current_dir not in valid_dirs:
        print("❌ 方向非法，跳过该段")
        continue
    
    # 4. 速度校验：超标/过低修正（你写）
    if current_speed > max_speed :
        current_speed = max_speed
        print(f"⚠️ 速度超标，修正为{max_speed}")
    elif current_speed< min_speed:
        current_speed = min_speed
        print(f"⚠️ 速度过低，修正为{min_speed}")
    
    # 5. 开灯规则：速度≥30且方向forward则开灯（你写）
    if current_speed >= 30 and current_dir == "forword":
        current_light = True
        light_on_count += 1
        print("✅ 速度≥30且正向行驶，打开车灯")
    
    # 6. 统计数据（你写）
    total_distance += current_dist
    speed_list.append(current_speed)
    if current_dir == "forward":
        forward_distance += current_dist
    
    # 7. 打印当前段有效信息
    print(f"✅ 方向：{current_dir} | 速度：{current_speed}km/h | 距离：{current_dist}米 | 车灯：{'开' if current_light else '关'}")
```
