```python
#初始化变量（小车初始速度）
current_speed = 20
#while + 条件： 只要条件成立，就重复执行缩进里的代码
while current_speed <40:
    current_speed  += 2 #每次加速2km/h
    print(f"当前速度: {current_speed}km/h") #打印当前速度
#条件不成立时（速度>=40, 跳出循环
print("已经达到目标速度，停止加速1")
```

```python
current_speed = 40
for i in range(5):
    current_speed -= 2
    print(f"第{i+1}次减速后，速度： {current_speed} km/h")
print("五次减速完成")
```

```python
# ====================== 第三课：循环结构（while/for）+ 小车调速实战 ======================
# 核心目标：掌握while/for循环，实现小车重复调速、自动启停

# ---------------------- 1. while循环：按条件重复（持续加速到目标速度） ----------------------
print("=== while循环：持续加速到目标速度 ===")
# 初始化参数
current_speed = 20    # 初始速度
target_speed = 40     # 目标速度
max_speed = 50        # 最大限速
speed_step = 2        # 每次加速步长

# while循环核心：条件成立就重复执行
while current_speed < target_speed:
    current_speed += speed_step  # 每次加速2km/h
    print(f"当前速度：{current_speed} km/h")
    
    # 限速保护：超过最大速度则强制终止循环
    if current_speed >= max_speed:
        print("⚠️ 即将超速，停止加速！")
        break  # break = 强制终止循环

# ---------------------- 2. for循环：按次数重复（固定次数减速） ----------------------
print("\n=== for循环：固定次数减速（5次）===")
# 重置速度
current_speed = 40

# for循环核心：range(5) = 执行5次（i从0到4）
for i in range(5):
    current_speed -= speed_step  # 每次减速2km/h
    
    # 跳过逻辑：速度低于10则跳过本次减速
    if current_speed < 10:
        print(f"⚠️ 速度过低（{current_speed}），跳过本次减速")
        current_speed += speed_step  # 恢复速度
        continue  # continue = 跳过本次循环，执行下一次
    
    # i+1：将「第0次」转为「第1次」，符合日常习惯
    print(f"第{i+1}次减速后速度：{current_speed} km/h")

# ---------------------- 3. 综合实战：循环+条件判断 → 小车人机交互调速 ----------------------
print("\n=== 综合实战：小车人机交互调速 ===")
# 初始化参数
current_speed = 15
car_running = True  # 小车运行状态（True=运行，False=停止）

print("🚗 小车启动！输入「加速」「减速」「停止」控制")
while car_running:
    # 接收用户指令（模拟传感器/按键输入）
    user_cmd = input("请输入指令：")
    
    # 条件判断处理指令
    if user_cmd == "加速":
        current_speed += 5
        print(f"✅ 加速后速度：{current_speed} km/h")
        # 限速检查
        if current_speed > max_speed:
            current_speed = max_speed
            print(f"⚠️ 已达最大限速：{max_speed} km/h")
    
    elif user_cmd == "减速":
        current_speed -= 5
        print(f"✅ 减速后速度：{current_speed} km/h")
        # 低速保护
        if current_speed < 0:
            current_speed = 0
            print(f"⚠️ 速度为0，小车静止")
    
    elif user_cmd == "停止":
        print("🛑 小车已停止！")
        car_running = False  # 终止while循环
    
    else:
        print("❌ 指令错误，请输入「加速」「减速」「停止」")

# ====================== 核心知识点速查 ======================
'''
1. while循环：while 条件: 执行逻辑 → 条件成立重复，不成立停止
2. for循环：for 变量 in range(次数): 执行逻辑 → 固定次数重复
3. 循环控制：
   - break：强制终止整个循环
   - continue：跳过本次循环，执行下一次
'''
```
