```python
# 树莓派小车基础变量练习 - 数据类型
#1. 整数（int) :小车速度（0-100）
car_speed = 50
#2. 字符串（str）: 行驶方向 （forword/ backword/left/right)
car_direction = "forword" 
#3.布尔值（bool） ：小车是否启动（True/False）
car_is_running = True
#4.浮点数（float） ： 小车电池电压
car_battery = 12.6

#打印变量值，查看类型
print("===小车基础参数===")
print(f"速度： {car_speed} | 类型： {type(car_speed)}")
print(f"方向： {car_direction} | 类型： {type(car_direction)}")
print(f"是否启动：{car_is_running} | 类型： {type(car_is_running)}")
print(f"电池电压：{car_battery} |类型 {type(car_battery)}")
```

```python
#练习1： 修改变量值
car_speed = 30 #降低速度
car_direction = "left" #切换方向
print (f"\修改后速度：{car_speed}, 方向：{car_direction}")

#练习2：类型转换（比如把输入的字符串速度转成整数）
input_speed = input("请输入小车目标速度（数字）：") #输入的是字符串
input_speed = int(input_speed) #转成整数
print(f"转换后速度： {input_speed} | 类型： {type(input_speed)}")

#练习3： 布尔值判断
if car_is_running:
    print("小车已启动 ")
else:
    print("小车未启动")
    
```

```python
### 知识点总结
1. Python基础数据类型：int（整数）、str（字符串）、bool（布尔值）、float（浮点数）；
2. 变量命名规则：小写+下划线，见名知意，不能以数字开头/含空格/用中文；
3. type()函数可以查看变量类型，input()输入的内容默认是字符串，需要手动转换。
```
