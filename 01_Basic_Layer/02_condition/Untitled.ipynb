```python
# 第二课：小车速度控制（变量运算+条件判断）
# 1. 定义基础变量（初始化小车参数）
car_speed = 30    # 初始速度（km/h）
max_speed = 50    # 最大限速（避免超速）
min_speed = 10    # 最低速度（避免停车）
speed_step = 5    # 每次加/减速的步长

# 2. 变量运算：模拟小车加速/减速
print("初始速度：", car_speed, "km/h")
car_speed += speed_step  # 加速5km/h（等价于car_speed = car_speed + 5）
print("加速后速度：", car_speed, "km/h")
car_speed *= 1.2         # 速度翻倍1.2倍（模拟超车）
print("超车后速度：", round(car_speed, 1), "km/h")  # 保留1位小数

# 3. 条件判断：根据速度自动调整行为
if car_speed > max_speed:
    # 超速：强制减速到限速
    car_speed = max_speed
    print(f"⚠️ 超速！已减速到最大限速：{car_speed} km/h")
elif car_speed < min_speed:
    # 速度过低：强制加速到最低速度
    car_speed = min_speed
    print(f"⚠️ 速度过低！已加速到最低速度：{car_speed} km/h")
else:
    # 速度正常：判断经济模式
    if 20 <= car_speed <= 40:
        print(f"✅ 速度正常（{car_speed} km/h），经济模式行驶")
    else:
        print(f"✅ 速度正常（{car_speed} km/h），正常模式行驶")

# 4. 拓展：取模运算判断速度奇偶（优化行驶稳定性）
if car_speed % 2 == 0:
    print(f"👍 当前速度{car_speed}是偶数，行驶更稳定")
else:
    car_speed += 1  # 微调为偶数
    print(f"🔧 微调速度到{car_speed}（偶数），提升稳定性")

# 最终结果
print("\n🚗 小车最终速度：", car_speed, "km/h")
```
